Data Cleaning And Preprocessing

```python
df
```

```python
df_encode=pd.get_dummies(df)
```

```python
df_encode
```

```python
df_encode=df_encode.astype(int)
```

```python
df_encode
```

```python
df_encode['ChestPainType_ATA'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
cols=['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']
scaler=StandardScaler()
df_encode[cols]=scaler.fit_transform(df[cols])
```

```python
df_encode
```

```python
X=df_encode.drop('HeartDisease',axis=1)
y=df_encode['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models={
    'logistic regression':LogisticRegression(),
    'knn':KNeighborsClassifier(),
    'naivebayes':GaussianNB(),
    'decisiontree':DecisionTreeClassifier(),
    'svm':SVC()
}
```

```python
results = []

for name, model in models.items():
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)

    results.append([name, acc])

results_df = pd.DataFrame(results, columns=['Model', 'Accuracy Score'])
results_df = results_df.sort_values(by='Accuracy Score', ascending=False)

results_df
```
